# RLT

```python
import os
import numpy as np
from IPython.display import Image
from IPython.display import display
import time
path = './metaworld_button-press-topdown-v2'
with open(path + '/_frame_idx.txt', 'r') as f:
    index = f.readlines()
index = [int(i) for i in index]
print(index)
```

# Human feedback

- If you prefer the segment above (press 1), equally prefer (press 2), or prefer the segment below (press 3)

- If you want to skip the segment pair (press 'n') or continue (press 'y')
    - Asks when a new segment is sampled.

- If you want to quit the program (input 'quit')

```python
kth = 0
seed = kth
import os
import numpy as np
from IPython.display import Image
from IPython.display import display
from IPython.display import clear_output
import time
curr = kth * 200
select = input('start: y,n')

sorted_query = []
feedback = 0
while feedback < 100:
    if len(sorted_query) == 0:
        l, r = index[curr], index[curr+1]
        curr += 2
        try:
            print(f'left: {l}, right: {r}')
            left = path + f'/render_{l}.gif'
            right = path + f'/render_{r}.gif'
            display(Image(filename=left))
            feedback += 1
            time.sleep(0.1)
            print('vs')
            display(Image(filename=right))
            select = input('select')
            if select == '1':
                sorted_query.append([r])
                sorted_query.append([l])
            elif select == '2':
                sorted_query.append([l, r])
            else:
                sorted_query.append([l])
                sorted_query.append([r])
        except:
            0 / 0
    else:
        clear_output()
        idx = index[curr]
        right = path + f'/render_{idx}.gif'
        display(Image(filename=right))
        time.sleep(0.1)
        select = input('select')
        if select == 'y':
            pass
        elif select == 'n':
            curr += 1
            continue
        try:
            low = 0
            high = len(sorted_query)
            pos = 0
            insert = False
            while low < high:
                clear_output()
                print(sorted_query)
                print('curr', curr)
                print('current idx', idx)
                print(f'{feedback} th feedback')
                mid = (low + high) // 2
                len_mid = len(sorted_query[mid])
                random_idx = np.random.randint(0, len_mid)
                compare_idx = sorted_query[mid][random_idx]
                print(f'left: {compare_idx}, right: {idx}')
                left = path + f'/render_{compare_idx}.gif'
                right = path + f'/render_{idx}.gif'
                display(Image(filename=left))
                time.sleep(0.1)
                print('vs')
                display(Image(filename=right))
                time.sleep(0.1)
                feedback += 1
                select = input('select')
                # exit the process
                if select == 'quit':
                    0 / 0
                elif select == '1':
                    high = mid
                    pos = mid
                elif select == '3':
                    low = mid + 1
                    pos = mid + 1
                else:
                    sorted_query[mid].append(idx)
                    insert = True
                    curr += 1
                    break
            if insert == False:
                sorted_query.insert(pos, [idx])
                curr += 1
        except:
            0 / 0
        with open(path + f'/list_label_{seed}.txt', 'w') as f:
            for i in range(len(sorted_query)):
                f.write(f'{i} {sorted_query[i]}\n')

# if already exist, do not save
if os.path.isfile(path + f'/list_label_{seed}.txt'):
    print('already exist')
else:           
    with open(path + f'/list_label_{seed}.txt', 'w') as f:
        for i in range(len(sorted_query)):
            f.write(f'{i} {sorted_query[i]}\n')
```
